##Pipeline

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("I've been waiting for a HuggingFace course my whole life.")
```

```python
classifier(
    ["I wanna retire and I am tired.", "I hate this so much!", "The future seems Bright and Glommy at the same time, but we will always be together."]
)
```

##Zero-Shot

```python
classifier = pipeline("zero-shot-classification")
classifier(
    ["This is a course about the Transformers library", "Metalurgy will be an important role for our Nation Economy","This game containing Violence, something that parents should be aware off"],
    candidate_labels=["education", "politics", "business"],
)
```

##Text-Generation

```python
generator = pipeline("text-generation")
generator("In this course, we will teach you how to")
```

##Other Model Uses

```python
generator = pipeline("text-generation", model="distilgpt2")
generator(
    "In this course, we will teach you how to",
    max_length=30,
    num_return_sequences=2,
)
```

##Mask-filling

```python
unmasker = pipeline("fill-mask")
unmasker("This course will teach you all about <mask> models.", top_k=2)
```

##Named-Entity-Recognition

```python
ner = pipeline("ner", grouped_entities=True)
ner("My name is Aditya and I study at Universitas Sriwijaya in Indralaya.")
```

##QnA System

```python
question_answerer = pipeline("question-answering")
question_answerer(
    question=["Where do you work?","Who are you?","Where do you Study","Where do you live?"],
    context="My name is Aditya and I study at Universitas Sriwijaya in Indralaya.",
)
```

#Transformer-Pipeline

```python
classifier = pipeline("sentiment-analysis")
classifier(
    [
        "I've been waiting for a HuggingFace course my whole life.",
        "I hate this so much!",
    ]
)
```

##Tokenizers

```python
from transformers import AutoTokenizer

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
```

```python
raw_inputs = [
    "I've been waiting for this course my whole life.",
    "I hate this so much!",
]
inputs = tokenizer(raw_inputs, padding=True, truncation=True, return_tensors="pt")
print(inputs)
```

##Models

```python
from transformers import AutoModel

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModel.from_pretrained(checkpoint)
```

```python
outputs = model(**inputs)
print(outputs.last_hidden_state.shape)
```

###Currently we are using sequence classification to separate sequence (positive or negative)

```python
from transformers import AutoModelForSequenceClassification

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)
outputs = model(**inputs)
```

```python
print(outputs.logits.shape)
```

```python
print(outputs.logits)
```

```python
import torch

predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)
print(predictions)
```
